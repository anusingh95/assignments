```python
import pandas as pd


def get_time(time):
    hours = time.seconds // 3600
    minutes = (time.seconds % 3600) // 60
    seconds = time.seconds % 60
    return hours, minutes,seconds


# Load the Excel file into a DataFrame
df = pd.read_excel("/home/anon/Downloads/Copy of Test data.xlsx")

# Convert 'date' column to datetime format
df['date'] = pd.to_datetime(df['date'])

# Convert 'time' column to timedelta format
df['time'] = pd.to_timedelta(df['time'].astype(str))

# Combine 'date' and 'time' columns to form 'datetime' column
df['datetime'] = df['date'] + df['time']

# Calculate the duration of each session
session_durations = df.groupby('session_id')['datetime'].apply(lambda x: x.max() - x.min())

# Task 1.1: Calculate the average number of questions asked per conversation for each day
avg_questions_per_day = df.groupby('date')['num_que_asked'].mean()
print("1.1 Average number of questions asked per conversation for each day:")
print(avg_questions_per_day)

# Task 1.2: Find out all the queries where the bot was not able to understand the query and assigned them to the agent
queries_assigned_to_agent = df[(df['is_bot'] == 'No') & (df['is_llm'] == 'No')]
print("\n1.2 Queries where the bot was not able to understand and assigned to the agent:")
print(queries_assigned_to_agent)

# Task 1.3: Calculate the average number of users talking to the agent for each day
avg_users_talking_to_agent = df[df['is_bot'] == 'No'].groupby('date')['user_id (new)'].nunique()
print("\n1.3 Average number of users talking to the agent for each day:")
print(avg_users_talking_to_agent)

# Task 1.4: Calculate the average time of conversation with the chat bot when the required information is provided by the chat bot
avg_time_when_info_provided = df[df['info_collection_end'] == 1].groupby('session_id')['datetime'].apply(lambda x: x.max() - x.min()).mean()

avg_time_hours, avg_time_minutes, avg_time_seconds = get_time(avg_time_when_info_provided)

print("\n1.4 Average time of conversation with the chat bot when the required information is provided by the chat bot: {} hours {} minutes {} seconds".format(avg_time_hours, avg_time_minutes, avg_time_seconds))


# Task 1.5: Calculate the average time of conversation with the chat bot when the required information is not provided by the chat bot and the user had to communicate with the agent
avg_time_when_info_not_provided = df[df['info_collection_end'] == 0].groupby('session_id')['datetime'].apply(lambda x: x.max() - x.min()).mean()

avg_time_not_provided_hours, avg_time_not_provided_minutes, avg_time_not_provided_seconds = get_time(avg_time_when_info_not_provided)

print("\n1.5 Average time of conversation with the chat bot when the required information is not provided by the chat bot and the user had to communicate with the agent: {} hours {} minutes {} seconds".format(avg_time_not_provided_hours, avg_time_not_provided_minutes, avg_time_not_provided_seconds))

```
